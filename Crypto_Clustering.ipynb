```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()
```

```python
# Create a DataFrame with the scaled data
df_market_scaled = pd.DataFrame(scaler.fit_transform(df_market_data), columns=df_market_data.columns)

# Copy the crypto names from the original data
crypto_names = df_market_data.index.to_list()

# Set the coinid column as index
df_market_scaled['coin_id'] = crypto_names
df_market_scaled = df_market_scaled.set_index('coin_id')

# Display sample data
df_market_scaled_sample = df_market_scaled.sample(10)
df_market_scaled_sample
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
k_values
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    # model = KMeans(n_clusters=k, random_state=0)
    model = KMeans(n_clusters=k, random_state=0, n_init=10)
    
    # 2. Fit the model to the data using `df_market_scaled`
    model.fit(df_market_scaled)
    
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)

inertia
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k_values, 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = df_elbow.hvplot.line(x='k', y='inertia', title='Elbow Curve', xlabel='Number of Clusters (k)', ylabel='Inertia', grid=True)

elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 
It seems the best value for `k` is where the elbow of the curve is located, which is probably 3 or 4 on the graph.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
# We'll start with k=3 for demonstration, but you can adjust based on your preference or further analysis
best_k = 3
kmeans_model = KMeans(n_clusters=best_k, random_state=0, n_init=10)

kmeans_model
```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(df_market_scaled)

# Returning a brief summary of the model
kmeans_model.labels_[:10]  # Displaying the cluster labels for the first 10 data points as a sample
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
predicted_clusters = kmeans_model.predict(df_market_scaled)

# Print the resulting array of cluster values.
predicted_clusters
```

```python
# Create a copy of the DataFrame
df_market_scaled_copy = df_market_scaled.copy()

df_market_scaled_copy.head()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_scaled_copy['predicted_cluster'] = predicted_clusters

# Display sample data
df_market_scaled_copy.sample(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

scatter_plot = df_market_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    hover_cols=["coin_id"],
    color="predicted_cluster",
    colormap="Category10",
    title="Cryptocurrencies Clustered by 24h and 7d Price Changes",
    xlabel="24h Price Change Percentage",
    ylabel="7d Price Change Percentage"
)

scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_data = pca.fit_transform(df_market_scaled_copy.drop(columns="predicted_cluster"))

# View the first five rows of the DataFrame.
df_pca = pd.DataFrame(
    data=pca_data, 
    columns=["principal component 1", "principal component 2", "principal component 3"],
    index=df_market_scaled_copy.index
)

df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

explained_variance
```

```python
# Create a new DataFrame with the PCA data.
df_pca_data = pd.DataFrame(pca_data, columns=["PC 1", "PC 2", "PC 3"])

# Copy the crypto names from the original data
crypto_names_pca = df_market_scaled_copy.index.tolist()

# Set the coinid column as index
df_pca_data['coin_id'] = crypto_names_pca
df_pca_data = df_pca_data.set_index('coin_id')

# Display sample data
df_pca_data.sample(10)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values_pca = list(range(1, 12))
k_values_pca
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values_pca:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    model_pca = KMeans(n_clusters=k, random_state=0, n_init=10)
    
    # 2. Fit the model to the data using `df_pca_data`
    model_pca.fit(df_pca_data)
    
    # 3. Append the model.inertia_ to the inertia list
    inertia_pca.append(model_pca.inertia_)

inertia_pca
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {'k': k_values_pca, 'inertia': inertia_pca}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data_pca)

df_elbow_pca
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot_pca = df_elbow_pca.hvplot.line(x='k', y='inertia', title='Elbow Curve for PCA Data', xlabel='Number of Clusters (k)', ylabel='Inertia', grid=True)

elbow_plot_pca
```

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
best_k_pca = 3
kmeans_model_pca = KMeans(n_clusters=best_k_pca, random_state=0, n_init=10)

kmeans_model_pca
```

```python
# Fit the K-Means model using the PCA data
kmeans_model_pca.fit(df_pca_data)

# Returning a brief summary of the model
kmeans_model_pca.labels_[:10]  # Displaying the cluster labels for the first 10 data points as a sample
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
predicted_clusters_pca = kmeans_model_pca.predict(df_pca_data)

# Print the resulting array of cluster values.
predicted_clusters_pca
```

```python
# Create a copy of the DataFrame with the PCA data
df_pca_data_copy = df_pca_data.copy()

# Add a new column to the DataFrame with the predicted clusters
df_pca_data_copy['predicted_cluster'] = predicted_clusters_pca

# Display sample data
df_pca_data_copy.sample(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scatter_plot_pca = df_pca_data_copy.hvplot.scatter(
    x="PC 1",
    y="PC 2",
    hover_cols=["coin_id"],
    color="predicted_cluster",
    colormap="Category10",
    title="Cryptocurrencies Clustered by PC1 and PC2",
    xlabel="Principal Component 1 (PC 1)",
    ylabel="Principal Component 2 (PC 2)"
)

scatter_plot_pca
```
